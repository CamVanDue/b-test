```python
from underthesea import pos_tag #tách từ
import pandas as pd
import math as m
import math
from normalization import *
import re

def score(q, d):
    ret = 0
    for t in q:
        ret += (tf(t, d) * idf(t))
    return ret

def keys_solve(string):
    string = string.lower()
    string = string.replace(",", " ").replace(".", " ")\
            .replace(";", " ").replace("“", " ") \
            .replace(":", " ").replace("”", " ") \
            .replace('"', " ").replace("'", " ") \
            .replace("!", " ").replace("?", " ") \
            .replace("[", " ").replace("]", " ").replace("'"," ")
    keys_before = pos_tag(string)    
    keys_filled=list()
    keys_after=dict()
    e = 0
    while (e < len(keys_before)): 
        keys_remove = keys_before[e]
        if keys_remove[1] not in ("E","L","Np","C"): 
            keys_filled.append(keys_remove[0])
        e += 1 
    keys_after["Input"] = keys_filled  
    keys_after = list(keys_after["Input"])
    long = len(keys_after)
    return keys_after

def tf(word,d):
    return d.count(word) /len(d)

def idf(word):
    count_in_document = 0 
    for d in D:
        if word in d:
            count_in_document +=1
    return 0 if count_in_document == 0 else math.log(len(D)/float(count_in_document))

def score(q, d):
    ret = 0
    for t in q:
        ret += (tf(t, d) * idf(t))
    return ret

#a) - Input 
Input = input("Nhập câu nào đó vào đây : ")
question_keys = keys_solve(Input) #VD như input là như vậy
#print(question_keys) #<--- Dùng lệnh này để kiểm tra khóa
query = question_keys
#b) - Word
df = pd.read_csv("./without_ques.txt",sep="/", names=["row"], encoding="utf8").dropna()
mt_df = df.as_matrix()
library_keys= dict()    
w = 0
q = 0
while (q<len(df)):   #từ 0 đến độ dài của dòng trong df ~ số câu trong bản word, có bao nhiêu câu sẽ có từng đó khóa
    line = str(mt_df[q])
    line = line.replace("["," ").replace("]"," ").replace("'"," ")
    keys_content = keys_solve(line)
    w+= 1
    key_name=  "cau"+str(w)    
    library_keys[key_name] = keys_content   
    q +=1
#print(library_keys) #Dùng lệnh này để kiểm tra khóa

D = library_keys.values()
D = list(D)

scores ={}
i=0
for d in D:
    #print(d)
    sc= score(query,d)
    if sc != 0:
        scores[i]=sc
    i+=1
#print(scores)

fp = open('./without_ques.txt','r',encoding="utf8" )
lis_doct=[line.replace(',', ' ') for line in fp.readlines()]
fp.close()

if len(scores)==0:
    print('Không có câu nào')
else:
    top_docs= sorted(scores, key=scores.get, reverse=True)[:40]
    rank=1
    for id in top_docs:
        print("Hạng : ",rank,'- ',scores[id])
        print(lis_doct[id].strip())
        print("-----------------------------------------------")
        rank+=1
        #print("Câu giống nhất : ", lis_doct[id].strip())
```
